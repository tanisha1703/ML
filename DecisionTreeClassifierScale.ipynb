```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('DataMidSem2024_2.csv')
```

```python
df1.head()
```

```python
X=df1.iloc[:,1:4]
X
```

```python
Y=df1['admit']
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
print("X_train shape",X_train.shape)
print("X_test shape",X_test.shape)
print("Y_train shape",Y_train.shape)
print("Y_test shape",Y_test.shape)
```

```python
admission=DecisionTreeClassifier()
```

```python
admission.fit(X_train,Y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(admission)
plt.show()
```

```python
admission_pred=admission.predict(X_test)
admission_pred
```

```python
score=accuracy_score(Y_test,admission_pred)
print("Accuracy Score:", score)
```

```python
cm1=confusion_matrix(Y_test,admission_pred)
print("Confusion matrix\n:",cm1)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay(cm1).plot()
plt.show()
```
